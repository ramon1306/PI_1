Resultado: contiene valores numéricos, se deben reemplazar.

```python
#1-Función para detectar los números
#2-Cambiar a string la columna
#3-Se aplica la función
#4-Ver el df luego de la condición

def tiene_num(cadena):
    return bool(re.search(r'\d', str(cadena)))

csv_movies_DS['original_language'] = csv_movies_DS['original_language'].astype(str)

fun_num = csv_movies_DS['original_language'].apply(tiene_num)

csv_movies_DS[fun_num].original_language
```

```python
#Cambiar números por NaN
csv_movies_DS.loc[fun_num, 'original_language'] = np.nan

#Comprobar
csv_movies_DS[fun_num].original_language 
```
